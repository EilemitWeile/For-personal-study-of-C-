```
%%writefile newton_method.c
// AI_Algorithm_202204637KimMinJae_HW2_Q4

#include <stdio.h>
#include <math.h>

#define MAX_ITER 100 // 최대 반복 횟수
#define THRESHOLD 1e-6 // 문턱값

// f1(x, y) = x^2 + y^2 - 4 = 0
// 함수 f1 정의: x^2 + y^2 - 4
double f1(double x, double y) {

    return x * x + y * y - 4.0;

}

// f2(x, y) = e^x + y - 1 = 0
// 함수 f2 정의: e^x + y - 1
double f2(double x, double y) {

    return exp(x) + y - 1.0;

}

//자코비안 1행 1열
// f1의 x에 대한 편미분: 2x
double df1_dx(double x) {

    return 2.0 * x;

}

//자코비안 1행 2열
// f1의 y에 대한 편미분: 2y
double df1_dy(double y) {

    return 2.0 * y;

}

//자코비안 2행 1열
// f2의 x에 대한 편미분: e^x
double df2_dx(double x) {

    return exp(x);

}

//자코비안 2행 2열
// f2의 y에 대한 편미분: 1
double df2_dy(double y) {

    return 1.0;

}


int main() {

    double x = 1; // x의 초기값
    double y = 1; // y의 초기값

    double F[2]; // 함수 벡터 F = [f1; f2]
    double J[2][2]; // 자코비안 행렬
    double detJ; // 자코비안의 행렬식
    double invJ[2][2]; // 자코비안의 역행렬
    double delta[2]; // 업데이트 벡터 delta = J^-1 * F

    int iter; // 반복 시행

    printf("비선형 뉴턴법 해\n");
    printf("초기값 : x = %.3f, y = %.3f\n", x, y);
    printf("Iter\t x\t\t y\t\t Error\n");

    // 뉴턴법
    for (iter = 1; iter <= MAX_ITER; iter++) {
        // 함수 벡터 F
        F[0] = f1(x, y);
        F[1] = f2(x, y);

        // F의 Euclidean Norm 사용해 수렴 경향 확인
        double error = sqrt(F[0]*F[0] + F[1]*F[1]);

        printf("%d\t %.3f\t %.3f\t %.3e\n", iter - 1, x, y, error);

        if (error < THRESHOLD) {
            printf("%% Convergence reached at iteration %d\n", iter - 1);
            break;
        }

        // 3. 자코비안 행렬 J 계산
        J[0][0] = df1_dx(x); // 2x
        J[0][1] = df1_dy(y); // 2y
        J[1][0] = df2_dx(x); // e^x
        J[1][1] = df2_dy(y); // 1

        // 자코비안의 행렬식 계산
        detJ = J[0][0] * J[1][1] - J[0][1] * J[1][0];

        if (fabs(detJ) < 1e-12) {
            printf("%% Error: Singular Jacobian matrix. Cannot invert.\n");
            return -1;
        }

        // 자코비안 역행렬 계산
        // J^-1 = (1/det) * [J22 -J12; -J21 J11]
        double invDet = 1.0 / detJ;
        invJ[0][0] =  J[1][1] * invDet;
        invJ[0][1] = -J[0][1] * invDet;
        invJ[1][0] = -J[1][0] * invDet;
        invJ[1][1] =  J[0][0] * invDet;

        // 업데이트 벡터 Delta = J^-1 * F
        delta[0] = invJ[0][0] * F[0] + invJ[0][1] * F[1];
        delta[1] = invJ[1][0] * F[0] + invJ[1][1] * F[1];

        // X_n = X_n-1 - Delta
        x = x - delta[0];
        y = y - delta[1];
    }


    if (iter > MAX_ITER) {
        printf("%% Warning: Maximum iterations reached without convergence.\n");
    }

    printf("\n%% Final Solution:\n");
    printf("x = %.3f\n", x);
    printf("y = %.3f\n", y);

    // 본식 대입해 성립 확인
    printf("\n%% Verification (f(x,y) should be close to 0):\n");
    printf("f1(x, y) = %.3f\n", f1(x, y));
    printf("f2(x, y) = %.3f\n", f2(x, y));

    return 0;
}
```
